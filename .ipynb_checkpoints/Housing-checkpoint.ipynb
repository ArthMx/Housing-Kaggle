# Missing Data

```python
null_values = data.isnull().sum().sort_values(ascending=False)
null_values[null_values!=0]
```

### Some data have nan values as a category ! Not because they are lacking the data ! 
Let's resolve that

```python
def nan_to_val(df):
    '''
    Change Nan values from categorical columns to an actual category value
    '''
    cols1 = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu',
           'GarageCond', 'GarageType', 'GarageFinish', 'GarageQual',
           'BsmtExposure', 'BsmtFinType2', 'BsmtFinType1', 'BsmtCond',
           'BsmtQual', 'MasVnrType']
    for col in cols1:
        df.loc[data[col].isnull(), col] = 'none'
    return df
```

```python
data = nan_to_val(data)
```

```python
null_values = data.isnull().sum().sort_values(ascending=False)
null_values[null_values!=0]
```

Assume that missing LotFrontage means no frontage : set nan to 0

Missing GarageYrBlt means no garage : set to 0

MasVnrArea assume that there is no massonery : set to 0

Electrical don't know if actual missing data or no electricity, let's drop this line because it's only one line.

```python
def nan_num_to_something(df):
    '''
    Replace Nan values of num columns
    '''
    df.loc[df.LotFrontage.isnull(), 'LotFrontage'] = 0
    
    df.loc[df.GarageYrBlt.isnull(), 'GarageYrBlt'] = 0
    
    df.loc[df.MasVnrArea.isnull(), 'MasVnrArea'] = 0
    
    return df
```

```python
data = nan_num_to_something(data)
data = data.dropna()
```

### No more Nan !
